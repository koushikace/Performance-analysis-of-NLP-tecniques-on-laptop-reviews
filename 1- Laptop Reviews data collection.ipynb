```python
  import pandas as reviews_of_lap_1

reviews_of_lap = reviews_of_lap_1.read_excel('Amazon laptop reviews.xlsx')
reviews_of_lap.shape
```

```python
reviews_of_lap = reviews_of_lap[['comment', 'rating']]
reviews_of_lap
```

```python
reviews_of_lap =reviews_of_lap.dropna()
```

**EDA**

```python
import seaborn as reviews_of_lap_2
import matplotlib.pyplot as reviews_of_lap_3
reviews_of_lap_3.figure(figsize=(10, 4))

reviews_of_lap_2.countplot(x ='rating', data = reviews_of_lap, palette = "Set2")
reviews_of_lap_3.title("Amazon Laptop Ratings details")
reviews_of_lap_3.show()
```

```python
reviews_of_lap['rating'].value_counts()
```

rating -2 is the minority class.

```python
from wordcloud import WordCloud as reviews_of_lap_4

lap_3=''.join([ff for ff in reviews_of_lap['comment']])

reviews_of_lap_3.figure(figsize=(10, 4))
wcd =reviews_of_lap_4(max_words=300, width=1000, height=400).generate(lap_3)
reviews_of_lap_3.imshow(wcd, interpolation='bilinear')
reviews_of_lap_3.axis('off')
reviews_of_lap_3.title("Laptop reviews content", fontsize=12)
reviews_of_lap_3.show()
```
